# ORBIT operator dashboard demo
Notebook-first dashboard flow for summary, navigation, approval drill-down, and state change review.

```python
from orbit.store import create_default_store
from orbit.runtime import SessionManager
from orbit.runtime.execution.backends import DummyExecutionBackend
from orbit.notebook import NotebookWorkbench
from orbit.settings import DEFAULT_WORKSPACE_ROOT

store = create_default_store()
session_manager = SessionManager(
    store=store,
    backend=DummyExecutionBackend(),
    workspace_root=str(DEFAULT_WORKSPACE_ROOT),
)
workbench = NotebookWorkbench(session_manager)

session_a = session_manager.create_session(backend_name='dummy', model='dummy-runtime')
session_b = session_manager.create_session(backend_name='dummy', model='dummy-runtime')
session_c = session_manager.create_session(backend_name='dummy', model='dummy-runtime')

plan_a = session_manager.run_dummy_session_turn(
    session_id=session_a.session_id,
    user_input='approval path',
    dummy_scenario='approval_required',
)
plan_b = session_manager.run_dummy_session_turn(
    session_id=session_b.session_id,
    user_input='final path',
    dummy_scenario='final_only',
)
plan_c = session_manager.run_dummy_session_turn(
    session_id=session_c.session_id,
    user_input='failure path',
    dummy_scenario='failure_case',
)
session_a.conversation_id, session_b.conversation_id, session_c.conversation_id
```

```python
waiting_df = workbench.open_approval_runs_dataframe()
waiting_df
```

```python
selected_run_id = waiting_df.iloc[0]['conversation_id']
workbench.run_summary_dataframe(selected_run_id)
```

```python
approval_df = workbench.approval_requests_for_run_dataframe(selected_run_id)
approval_df
```

```python
approval_id = approval_df.iloc[0]['approval_request_id']
resolved_plan = workbench.approve(approval_id, 'approved from operator dashboard demo')
resolved_plan.plan_label, resolved_plan.final_text
```
