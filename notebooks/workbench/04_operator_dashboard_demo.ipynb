# ORBIT operator dashboard demo
Notebook-first dashboard flow for summary, navigation, approval drill-down, and state change review.

```python
from orbit.store import create_default_store
from orbit.runtime import OrbitCoordinator
from orbit.notebook import NotebookWorkbench
from orbit.settings import DEFAULT_WORKSPACE_ROOT

store = create_default_store()
coordinator = OrbitCoordinator(store, DEFAULT_WORKSPACE_ROOT)
workbench = NotebookWorkbench(coordinator)

run_a = coordinator.run_dummy_scenario(
    title='Dashboard demo A',
    description='approval-required run',
    user_input='approval path',
    dummy_scenario='approval_required',
)
run_b = coordinator.run_dummy_scenario(
    title='Dashboard demo B',
    description='completed run',
    user_input='final path',
    dummy_scenario='final_only',
)
run_c = coordinator.run_dummy_scenario(
    title='Dashboard demo C',
    description='failed run',
    user_input='failure path',
    dummy_scenario='failure_case',
)
run_a.run_id, run_b.run_id, run_c.run_id
```

```python
waiting_df = workbench.open_approval_runs_dataframe()
waiting_df
```

```python
selected_run_id = waiting_df.iloc[0]['run_id']
workbench.run_summary_dataframe(selected_run_id)
```

```python
approval_df = workbench.approval_requests_for_run_dataframe(selected_run_id)
approval_df
```

```python
approval_id = approval_df.iloc[0]['approval_request_id']
resolved_run = workbench.approve(approval_id, 'approved from operator dashboard demo')
resolved_run.run_id, resolved_run.status
```
