# ORBIT run set analysis demo
Notebook workbench surface for task/run listings and multi-run event analysis.

```python
from orbit.store import create_default_store
from orbit.runtime import SessionManager
from orbit.runtime.execution.backends import DummyExecutionBackend
from orbit.notebook import NotebookWorkbench
from orbit.settings import DEFAULT_WORKSPACE_ROOT

store = create_default_store()
session_manager = SessionManager(
    store=store,
    backend=DummyExecutionBackend(),
    workspace_root=str(DEFAULT_WORKSPACE_ROOT),
)
workbench = NotebookWorkbench(session_manager)

session_a = session_manager.create_session(backend_name='dummy', model='dummy-runtime')
session_b = session_manager.create_session(backend_name='dummy', model='dummy-runtime')

plan_a = session_manager.run_dummy_session_turn(
    session_id=session_a.session_id,
    user_input='Run an approval-required scenario.',
    dummy_scenario='approval_required',
)
plan_b = session_manager.run_dummy_session_turn(
    session_id=session_b.session_id,
    user_input='Run a failure scenario.',
    dummy_scenario='failure_case',
)
session_a.session_id, session_b.session_id, plan_a.plan_label, plan_b.plan_label
```
