# ORBIT SSH vLLM Hello World Demo

This notebook demonstrates the first narrow SSH vLLM comparison path for ORBIT.


```python
from pathlib import Path

from orbit.notebook.vllm_demo import (
    run_ssh_vllm_hello_world,
    ssh_vllm_hello_world_summary_frame,
)

WORKSPACE_ROOT = Path.cwd().resolve().parent if Path.cwd().resolve().name == 'notebooks' else Path.cwd().resolve()
USER_INPUT = 'Hello from ORBIT. Please reply with a short hello-world style confirmation.'
REMOTE_BASE_URL = 'http://127.0.0.1:8000/v1'
MODEL = 'Qwen/Qwen2.5-Coder-14B-Instruct-AWQ'
API_KEY = 'EMPTY'
AUTO_TUNNEL = True
SSH_HOST = 'ws3'
```

## Execute first SSH vLLM hello-world call

If `AUTO_TUNNEL = True`, ORBIT will temporarily establish `ssh -L` forwarding through the configured SSH host for the duration of the request.


```python
result = run_ssh_vllm_hello_world(
    user_input=USER_INPUT,
    workspace_root=WORKSPACE_ROOT,
    remote_base_url=REMOTE_BASE_URL,
    model=MODEL,
    api_key=API_KEY,
    timeout_seconds=60,
    auto_tunnel=AUTO_TUNNEL,
    ssh_host=SSH_HOST,
    local_port=8000,
    remote_host='127.0.0.1',
    remote_port=8000,
)
summary_df = ssh_vllm_hello_world_summary_frame(result)
summary_df
```
