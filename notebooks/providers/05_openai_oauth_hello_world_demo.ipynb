# ORBIT OpenAI Codex Hello World Demo

This notebook demonstrates ORBIT's first notebook-driven OpenAI Codex hello-world path using a persisted local OAuth credential.


```python
from pathlib import Path

from orbit.notebook import (
    create_openai_login_url_bundle,
    openai_codex_hello_world_summary_frame,
    openai_login_url_summary_frame,
    run_openai_codex_hello_world,
)

REPO_ROOT = Path.cwd().resolve().parent if Path.cwd().resolve().name == 'notebooks' else Path.cwd().resolve()
USER_INPUT = 'Hello from ORBIT. Please reply with a short hello-world style confirmation.'
MODEL = 'gpt-5.4'
API_BASE = 'https://chatgpt.com/backend-api'
```

## Generate OpenAI login URL

Run this cell if you need a fresh browser login URL before persisting the OAuth credential.


```python
login_bundle = create_openai_login_url_bundle(repo_root=REPO_ROOT)
login_df = openai_login_url_summary_frame(login_bundle)
login_df
```

```python
print(login_bundle['authorize_url'])
```

## Execute first OpenAI Codex hello-world call

This uses the persisted credential stored under `ORBIT/.runtime/openai_oauth_credentials.json`.

If the notebook kernel was started inside the `notebooks/` directory, the repo root is automatically normalized back to the ORBIT root before loading the credential.


```python
result = run_openai_codex_hello_world(
    user_input=USER_INPUT,
    repo_root=REPO_ROOT,
    model=MODEL,
    api_base=API_BASE,
    timeout_seconds=60,
)
summary_df = openai_codex_hello_world_summary_frame(result)
summary_df
```
