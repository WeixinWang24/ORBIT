# ORBIT v0 demo
Notebook-first workbench surface for Phase 0/1 experimentation.

```python
from orbit.store import create_default_store
from orbit.runtime import SessionManager
from orbit.runtime.execution.backends import DummyExecutionBackend
from orbit.notebook import project_run, events_dataframe, session_messages_dataframe
from orbit.settings import DEFAULT_WORKSPACE_ROOT

store = create_default_store()
session_manager = SessionManager(
    store=store,
    backend=DummyExecutionBackend(),
    workspace_root=str(DEFAULT_WORKSPACE_ROOT),
)
session = session_manager.create_session(backend_name='dummy', model='dummy-runtime')
plan = session_manager.run_dummy_session_turn(
    session_id=session.session_id,
    user_input='Create a small file through a governed tool path.',
    dummy_scenario='tool_then_finish',
)
run_id = session.conversation_id
session.session_id, run_id, plan.plan_label
```

```python
session_messages_dataframe(store, session.session_id)
```

```python
events_dataframe(store, run_id)
```

```python
project_run(store, run_id)
```
